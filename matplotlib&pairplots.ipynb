```python
import seaborn as sns; sns.set(style="ticks", color_codes=True)
iris = sns.load_dataset("iris")
g = sns.pairplot(iris, hue="species", palette="husl", markers=["o", "s", "D"], kind="reg")
```

```python
g=sns.pairplot(data=iris, vars=["petal_length", "petal_width"], hue="species", markers=["o", "s", "D"], diag_kind="hist", height=4)
g.fig.suptitle("Pairplots", y=1.05, size=20, color='r')
```

```python
import numpy as np, pandas as pd; np.random.seed(0)
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style="white", color_codes=True)
tips = sns.load_dataset("tips")
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind='reg')
g.fig.suptitle("Jointplot", y=1.05, size=20, color='b')
plt.savefig(r'C:\Users\jithi\Desktop\jp.png', dpi=400)
```

```python
plt.style.use("ggplot")
fig, ax=plt.subplots(2,1, sharey=True, sharex=True)
ax[0].plot(tips.tip, marker="o", alpha=.7, color='g')
ax[0].set_ylabel("Hello", size=15)
ax[0].legend("Green")
ax[1].plot(tips.total_bill, marker="v", alpha=.7, linestyle='--', color='m')
ax[1].set_ylabel("hi", size=15)
ax[1].set_xlabel("hi", size=15)
ax[1].legend("Magenta")
fig.suptitle("Plot", size=25)
fig.set_size_inches([10,5])
fig.savefig(r'C:\Users\jithi\Desktop\plot.png', dpi=600)
```
